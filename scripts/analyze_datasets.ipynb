```python
import numpy as np
import pandas as pd
import pyterrier as pt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vaswani_dataset = pt.datasets.get_dataset("vaswani")
antique_dataset = pt.datasets.get_dataset("irds:antique/test/non-offensive")
lotte_dataset = pt.datasets.get_dataset("irds:lotte/lifestyle/dev/forum")

vaswani_topics = vaswani_dataset.get_topics()
antique_topics = antique_dataset.get_topics()
lotte_topics = lotte_dataset.get_topics()
```

```python
def analyze_datasets(datasets):
    for dataset, topics in datasets.items():
        query_term_counts = {}
        for qid in topics.index:
            token_count = len(topics.loc[qid, 'query'].split())
            query_term_counts[qid] = token_count

        min_terms = min(query_term_counts.values())
        max_terms = max(query_term_counts.values())

        sns.set_style("whitegrid")
        plt.figure(figsize=(10, 6))

        bins = np.arange(min_terms - 0.5, max_terms + 1.5, 1)
        ax = sns.histplot(query_term_counts.values(), bins=bins, facecolor='#009FFF', edgecolor='white', alpha=1)
        sns.despine(left=True)

        plt.xticks(range(min_terms, max_terms + 2))
        plt.xlabel('Number of Terms')
        plt.ylabel('Number of Queries')
        plt.title(f'Distribution of Terms per Query in {dataset} Dataset')
        ax.grid(True, axis='y')
        ax.grid(False, axis='x')
        ax.set_facecolor('#f0f0f4')
        ax.legend_.remove()
        plt.xticks(range(min_terms, max_terms + 1))

        plt.tight_layout()
        plt.savefig(f"plots/term_distribution_{dataset}.png", dpi=300, bbox_inches='tight')
        plt.show()
```

```python
datasets = {
    "Vaswani": vaswani_topics,
    "Antique": antique_topics,
    "Lotte": lotte_topics
}

analyze_datasets(datasets)
```
